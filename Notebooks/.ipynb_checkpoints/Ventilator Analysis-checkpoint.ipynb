# Compute the Range of Drone
The following python notebook computes the range of delivery drone from property values defined in the Drone model. The Drone model is load in the following code.s.

```python
import capellambse

from IPython.core.display import HTML
from IPython.core.display import HTML

resources = {
    "AIDA Siemens Hobby Propulsion": "AIDA_Siemens_Hobby_Propulsion/AIDA Siemens Hobby Propulsion",
}
path_to_model = "AIDA_Siemens_Hobby_Drone_System/AIDA Siemens Hobby Drone System/AIDA_hobby_drone.aird"
model = capellambse.MelodyModel(path_to_model, resources=resources, jupyter_untrusted=False)
```

Load up the relevent properties from the model.

```python
from IPython import display
display.Image("AIDA_Siemens_Hobby_Drone_System/AIDA Siemens Hobby Drone System/Python4Capella_exported_diagrams/[LCBD] Drone Range.jpg")
```

```python
drone = model.search("LogicalComponent").by_name("Drone")

print(drone.property_value_groups["Goal Payload"]["Value"])
```

```python
def Roll_Up(Object,Propery_Name,Units,Sum) :
    for Object in Object.components:
        Sum = Roll_Up(Object,Propery_Name,Units,Sum) 
        if Object.property_value_groups[Propery_Name]["Units"] == Units :
            print("Adding", Object.name, "Value:", Object.property_value_groups[Propery_Name]["Value"] , "to Roll Up Sum",Sum)
            Sum = Sum + float(Object.property_value_groups[Propery_Name]["Value"])
        else :
            print("Value for", Object.name, "is not the unit",Units, "but is" ,Object.property_value_groups[Propery_Name]["Units"])
    return Sum
```

Roll Up Mass of Drone of Drone Subsysems and Assing it Drone and save the value in Model. 

```python
#help(drone)
print(drone.name)
drone.property_value_groups["Mass"]["Value"] = Roll_Up(drone,"Mass","Kg",0.0)
model.save()
```

```python
print("Drone mass", drone.property_value_groups["Mass"]["Value"])
```

Load up the values for the Payload and Battery size to run a calculation of range. 

```python
Goal_Payload = drone.property_value_groups["Goal Payload"]["Value"]
Battery_Size =  model.search("LogicalComponent").by_name("Power System").property_value_groups["Min_Power"]["Value"]
print("Goal_Payload:",Goal_Payload,' ',drone.property_value_groups["Goal Payload"]["Units"])
print("Battery_Size:",Battery_Size,' ',model.search("LogicalComponent").by_name("Power System").property_value_groups["Min_Power"]["Units"])
```

Compute the range over a typical flight profile. 

```python
Range = (Battery_Size/Goal_Payload ) / 1000
print("Range",Range)
```

Compare the range to requirement and determine if requirement is met.

```python
if Range >= drone.property_value_groups["Minimum_Range"]["Value"] :
    print("Range Requirement Passes verification. Estimated:",Range, "Requirement:",drone.property_value_groups["Minimum_Range"]["Value"])
else :
    print("Range Requirement Fails verification. Estimated:",Range, "Requirement:",drone.property_value_groups["Minimum_Range"]["Value"])
```
